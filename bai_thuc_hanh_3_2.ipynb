Thực hiện tìm các contour trong ảnh và xác định đối tượng

```python
import numpy as np
import cv2 as cv 
import matplotlib.pyplot as plt
```

```python
img = cv.imread('spade.png')
#Convert to grayscale
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(img)
plt.show()
```

```python
ret, thresh = cv.threshold(img_gray, 127, 255, 0) # convert to binary image
contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
plt.imshow(thresh)
plt.show()
```

```python
# Vẽ đường viền
# Draw all contours 
# -1 signifies drawing all contours
cv.drawContours(img, contours, -1, (0,255,0), 5) # 5 mean that offset = 5
plt.imshow(img)
plt.show()
```

```python
# Vẽ bounding rectangle cho các đường viền
#cnt = contours[3]
for i in range(len(contours)):
    x,y,w,h = cv.boundingRect(contours[i])
    cv.rectangle(img,(x,y),(x+w,y+h),(255,0,0),5)
plt.imshow(img)
plt.show()
```
