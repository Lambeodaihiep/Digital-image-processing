Bài toán xác định đối tượng (Object detection) và tìm tọa độ của đối tượng trong ảnh:
- Pattern/Template Matching (PM)
    Đem 1 cái template m*n đi duyệt vào ảnh
    So sánh độ tương quan giữa từng vùng trong ảnh và mẫu cần tìm
- Object dectection sử dụng findContour

```python
import numpy as np
import cv2 as cv 
import matplotlib.pyplot as plt
```

```python
img = cv.imread('duck.jpeg')
#Convert to grayscale
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
template = cv.imread('template.jpg', 0)
w, h = template.shape[1], template.shape[0]
plt.imshow(img)
plt.show()
```

```python
res = cv.matchTemplate(img_gray,template, cv.TM_CCOEFF_NORMED)
plt.imshow(res)
plt.show()
```

```python
THRESHOLD = 0.65
loc = np.where(res >= THRESHOLD)

#Draw boudning box
for y, x in zip(loc[0], loc[1]):
    cv.rectangle(img, (x, y), (x + w, y + h), (255,0,0), 1)
plt.imshow(img)
plt.show()
```
