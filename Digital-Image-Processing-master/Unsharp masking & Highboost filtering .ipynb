```python
#image sharpening using unsharp masking 

import cv2
import matplotlib.pyplot as plt 
import math
import numpy as np

	
# Read the image 
img = cv2.imread('images/coin.jpg', 0) 

# Obtain number of rows and columns 
# of the image 
m, n = img.shape 

# Develop Averaging filter(3, 3) mask 
mask = np.ones([3, 3], dtype = int) 
mask = mask / 9

# Convolve the 3X3 mask over the image for blurring 
img_new = np.zeros([m, n]) 

for i in range(1, m-1): 
	for j in range(1, n-1): 
		temp = img[i-1, j-1]*mask[0, 0]+img[i-1, j]*mask[0, 1]+img[i-1, j + 1]*mask[0, 2]+img[i, j-1]*mask[1, 0]+ img[i, j]*mask[1, 1]+img[i, j + 1]*mask[1, 2]+img[i + 1, j-1]*mask[2, 0]+img[i + 1, j]*mask[2, 1]+img[i + 1, j + 1]*mask[2, 2] 
		
		img_new[i, j]= temp 
		
img_new = img_new.astype(np.uint8) 

img_mask = np.zeros([m, n]) 
img_mask = img - img_new 

#formula for unsharp masking : sharped image = orignal image + (original image - blurred image)

img_new = img + img_mask 



# Display the original image
plt.imshow(img) 
plt.show()

# Display the mask
plt.imshow(img_mask) 
plt.show()

# Display the sharpened image
plt.imshow(img_new) 
plt.show()
```

```python
#image sharpening using highboos filtering  
'''
formula for unsharp masking : 
sharped image = orignal image + k*(original image - blurred image)
    where k = 1
if k > 1 it is highboost filtering 

'''

import cv2
import matplotlib.pyplot as plt 
import math
import numpy as np

	
# Read the image 
img = cv2.imread('images/coin.jpg', 0) 

k = 2

# Obtain number of rows and columns 
# of the image 
m, n = img.shape 

# Develop Averaging filter(3, 3) mask 
mask = np.ones([3, 3], dtype = int) 
mask = mask / 9

# Convolve the 3X3 mask over the image for blurring 
img_new = np.zeros([m, n]) 

for i in range(1, m-1): 
	for j in range(1, n-1): 
		temp = img[i-1, j-1]*mask[0, 0]+img[i-1, j]*mask[0, 1]+img[i-1, j + 1]*mask[0, 2]+img[i, j-1]*mask[1, 0]+ img[i, j]*mask[1, 1]+img[i, j + 1]*mask[1, 2]+img[i + 1, j-1]*mask[2, 0]+img[i + 1, j]*mask[2, 1]+img[i + 1, j + 1]*mask[2, 2] 
		
		img_new[i, j]= temp 
		
img_new = img_new.astype(np.uint8) 

img_mask = np.zeros([m, n]) 
img_mask = img - img_new 

#formula for unsharp masking : sharped image = orignal image + (original image - blurred image)

img_new = img + k * img_mask 



# Display the original image
plt.imshow(img) 
plt.show()

# Display the mask
plt.imshow(img_mask) 
plt.show()

# Display the sharpened image
plt.imshow(img_new) 
plt.show()
```
